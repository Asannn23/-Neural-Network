```python
# Imports
import pandas as pd
from pathlib import Path
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
```

```python
applicant_data_df = pd.read_csv("applicants_data.csv")

applicant_data_df.head()
```

```python
# Drop the 'EIN' and 'NAME' columns from the DataFrame
applicant_data_df =  applicant_data_df.drop(columns = ["EIN", "NAME"])
```

```python
# review the dataframe
applicant_data_df.head()
```

```python
# Create a OneHotEncoder instance

enc = OneHotEncoder(sparse = False)
```

```python
list(applicant_data_df.dtypes[applicant_data_df.dtypes =="object"].index)
```

```python
categorical_variables = ['APPLICATION_TYPE',
 'AFFILIATION',
 'CLASSIFICATION',
 'USE_CASE',
 'ORGANIZATION',
 'INCOME_AMT',
 'SPECIAL_CONSIDERATIONS']
```

```python
encoded_data = enc.fit_transform(applicant_data_df[categorical_variables])
```

```python
# Create DataFrame with encoded variables

encoded_df = pd.DataFrame(
    encoded_data,
    columns = enc.get_feature_names(categorical_variables)
)

encoded_df.head()
```

```python
numerical_variables_df = applicant_data_df.drop(columns = categorical_variables)

numerical_variables_df.head()
```

```python
# combining two dataframe into one

encoded_df = pd.concat(
    [
        numerical_variables_df,
     encoded_df
    ],
    axis = 1
)
encoded_df.head()
```

```python
y = encoded_df["IS_SUCCESSFUL"]

y[:5]
```

```python

X = encoded_df.drop(columns=["IS_SUCCESSFUL"])

# Review the features DataFrame
X.head()
```

```python
X.shape
```

```python
# Split the data into training and testing datasets
# Assign the function a random_state equal to 1
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the scaler to the features training dataset
X_scaler = scaler.fit(X_train)

# Fit the scaler to the features training dataset
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python

number_input_features = len(X_train.iloc[0])


hidden_nodes_layer1 =  (number_input_features + 1) // 2 


hidden_nodes_layer2 = (hidden_nodes_layer1 + 1) // 2


nn = Sequential()


nn.add(Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu"))


nn.add(Dense(units=hidden_nodes_layer2, activation="relu"))


nn.add(Dense(units=1, activation="sigmoid"))
```

```python
nn.summary()
```

```python
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python

fit_model = nn.fit(X_train_scaled, y_train, epochs=50)
```

```python
# Evaluate the model loss and accuracy metrics using the evaluate method and the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)

# Display the model loss and accuracy results
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# bulding the first model 

# Import the `Dropout` layer
from keras.layers import Dropout
# Import regularization functions
from keras.regularizers import l1, l2
```

```python
number_of_predictors = len(X.columns)
len(X.columns)
```

```python
# Build the Model
nn_A1= Sequential()
# Add a dense layer
nn_A1.add(Dense(10, input_dim=number_of_predictors, activation='relu'))
# Now add a dropout layer
nn_A1.add(Dropout(.2,input_shape=(10,)))
# Add another dense layer
nn_A1.add(Dense(5, activation='relu', kernel_regularizer=l2(0.01), bias_regularizer=l2(0.01)))
```

```python
nn_A1.add(Dense(1, activation= "softmax"))
```

```python
# Compile the model
nn_A1.compile(loss="categorical_crossentropy", 
              optimizer= "adam", 
              metrics=['categorical_accuracy'])
```

```python
# Fit the model
number_of_epochs = 50
nn_A1.fit(X_train,y_train, 
                    epochs=number_of_epochs,
                    batch_size=10,
                      validation_split=0.2,
                    shuffle=True)
```

```python
# Evaluate the model loss and accuracy metrics using the evaluate method and the test data
model_A1_loss, model_A1_accuracy = nn_A1.evaluate(X_test_scaled,y_test,verbose=2)

# Display the model loss and accuracy results
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

SECOND MODEL:

```python

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, random_state=78)
```

```python

from sklearn.preprocessing import StandardScaler


scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
```

```python
# Model set-up
number_input_features = 10
hidden_nodes_layer1 = 15
hidden_nodes_layer2 = 5
```

```python
# Define the LSTM RNN model
nn_A2 = Sequential()

# Layer 1
nn_A2.add(
    Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)

# Layer 2
nn_A2.add(Dense(units=hidden_nodes_layer2, activation="relu"))

# Output layer
nn_A2.add(Dense(1, activation="sigmoid"))
```

```python
# Compile the model
nn_A2.compile(
    loss="binary_crossentropy",
    optimizer="adam",
    metrics=[
        "accuracy",
        tf.keras.metrics.TruePositives(name="tp"),
        tf.keras.metrics.TrueNegatives(name="tn"),
        tf.keras.metrics.FalsePositives(name="fp"),
        tf.keras.metrics.FalseNegatives(name="fn"),
        tf.keras.metrics.Precision(name="precision"),
        tf.keras.metrics.Recall(name="recall"),
        tf.keras.metrics.AUC(name="auc"),
    ],
)
```

```python

batch_size = 1000
epochs = 50
training_history = nn_A2.fit(
    X_train,
    y_train,
    validation_data=(X_val, y_val),
    epochs=epochs,
    batch_size=batch_size,
    verbose=1
)
```
